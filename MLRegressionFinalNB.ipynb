```python
from sklearn.naive_bayes import GaussianNB
import pandas as pd
import csv
from sklearn.cross_validation import train_test_split
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
df = pd.read_csv("regLin.csv")
df
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(df[["X"]],df[["y"]], train_size=0.75)
```

```python
clf = LinearRegression()
clf
clf.fit(X_train, Y_train)
```

```python
np.mean((clf.predict(X_test)-Y_test) ** 2)
```

```python
clf.coef_, clf.intercept_
```

```python
plt.scatter(X_test,Y_test,color='blue')
plt.plot(X_test,clf.predict(X_test),color='red')
```

```python
w0= clf.intercept_[0]
w1=clf.coef_[0][0]

W=range(-5,5,1)+w0
error=[]
for i in W:
    y_predecida=w1*X_test+i
    C=np.subtract(y_predecida, Y_test)
    C=C**2
    error.append(np.mean(C))
plt.plot(W,error,color="red")
```
